```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd

df = pd.read_csv('movies_metadata.csv')
df.head()
```

```python
df.columns

colNeeded= [
    'genres',
    'original_title',
    'overview','tagline'
]

df_rec = df[colNeeded]
df_rec.head()
```

```python
# df_rec['genres'][0]

# df_rec.drop_duplicates()
df_rec = df_rec.fillna("")

df_rec.isnull().sum()
```

```python
import ast

df_rec["genres"] = df_rec["genres"].apply(
    lambda x: " ".join(
        d["name"] for d in ast.literal_eval("".join(x))
    )
)

df_rec.head()
```

```python
df_rec["keywords"] = (
    df_rec["genres"] + " " +
    df_rec["original_title"] + " " +
    df_rec["overview"] + " " +
    df_rec["tagline"]
)

df_rec_dl = df_rec.copy()

df_rec = df_rec[["keywords", "original_title"]]
df_rec.rename(columns={"original_title": "title"}, inplace=True)

df_rec.head()
```
