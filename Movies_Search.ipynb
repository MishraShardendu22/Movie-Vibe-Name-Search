```python
import pandas as pd

df = pd.read_csv('movies_metadata.csv')
df.head()
```

```python
df.columns

colNeeded= [
    'genres',
    'original_title',
    'overview','tagline'
]

df_rec = df[colNeeded]
df_rec.head()
```

```python
# df_rec['genres'][0]

# df_rec.drop_duplicates()
df_rec = df_rec.fillna("")

df_rec.isnull().sum()
```

```python
import ast

df_rec["genres"] = df_rec["genres"].apply(
    lambda x: " ".join(
        d["name"] for d in ast.literal_eval("".join(x))
    )
)

df_rec.head()
```

```python
df_rec["keywords"] = (
    df_rec["genres"] + " " +
    df_rec["original_title"] + " " +
    df_rec["overview"] + " " +
    df_rec["tagline"]
)

df_rec_dl = df_rec.copy()

df_rec = df_rec[["keywords", "original_title"]]
df_rec.rename(columns={"original_title": "title"}, inplace=True)

df_rec.head()
```

# ML - NLP Search

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download("stopwords")
nltk.download("wordnet")

stop_words = set(stopwords.words("english"))
roman = {"i","ii","iii","iv","v","vi","vii","viii","ix","x"}
lemmatizer = WordNetLemmatizer()
```

```python
def sentenceCleansing(sen):
    sen = str(sen).lower()
    sen = re.sub(r'[^a-z\s]', '', sen)
    words = sen.split()
    words = [w for w in words if w not in stop_words and w not in roman]
    words = [lemmatizer.lemmatize(w) for w in words]
    return " ".join(words)

df_rec["keywords"] = df_rec["keywords"].apply(sentenceCleansing)
df_rec.head()
```

```python
# TF-IDF = Term Frequency â€“ Inverse Document Frequency
# Term Frequency (TF): how often a word appears in a document.
# Inverse Document Frequency (IDF): how rare that word is across all documents.

from sklearn.feature_extraction.text import TfidfVectorizer

indices = pd.Series(df_rec.index, index=df_rec["title"]).drop_duplicates()
indices
```

```python
tfidf = TfidfVectorizer(
    max_features=5000000,
    ngram_range=(1,2),
    stop_words="english"
)

tfidf_matrix = tfidf.fit_transform(df_rec["keywords"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import difflib

def recommend(title, n=10):
    if title not in indices:
        matches = difflib.get_close_matches(title, indices.index, n=1, cutoff=0.6)
        if not matches:
            return ["Movie not found"]
        title = matches[0]

    idx = indices[title]
    sim = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
    sim_idx = sim.argsort()[::-1][1:n+1]
    return df_rec["title"].iloc[sim_idx]
```

```python
recommend("Avatar", 5)
```

# DL - NLP Search

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-mpnet-base-v2")

embeddings = model.encode(
    df_rec["keywords"].tolist(),
    show_progress_bar=True
)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_dl(query, n=10):
    q_emb = model.encode([query])
    sim = cosine_similarity(q_emb, embeddings).flatten()
    sim_idx = sim.argsort()[::-1][:n]
    return df_rec["title"].iloc[sim_idx]
```

```python
recommend_dl("humans invade alien world for resources sci fi movie about exploitation of nature", 15)
```
